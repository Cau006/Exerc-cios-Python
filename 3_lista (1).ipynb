# 5. Pygame - Capturando eventos do teclado

## Desenvolva um minijogo onde o usuário deve pressionar a tecla espaço para fazer um personagem pular. Implemente a detecção dessa tecla para iniciar o salto.

## Bibliotecas/Métodos: pygame.KEYDOWN, pygame.K_SPACE

```python
import pygame
import sys
from sys import exit
from pygame.locals import *


pygame.init()

largura = 600
altura = 600

x = largura/2
y = altura/2

janela = pygame.display.set_mode((largura,altura))
pygame.display.set_caption('Elon Musk')
relogio = pygame.time.Clock()

while True:
    relogio.tick(60)
    janela.fill((0,0,0))
    for event in pygame.event.get():
        if event.type == QUIT:
            pygame.quit()
            exit()
        if event.type == KEYDOWN:
            if event.key == K_SPACE: 
                y = y - 60
                
            elif event.key == K_d:     
                x = x + 60
            
            elif event.key == K_a:
                x = x - 60
            
            elif event.key == K_s:
                y = y + 60
                
            elif event.key == K_w:
                y = y - 60
            
            
            
    
    personagem = pygame.draw.rect(janela, (0, 255 , 0 ), (x, y, 50 , 50))
    pygame.display.update()
```

# 7. OpenCV (cv2) - Convertendo imagens para escala de cinza

## Você está trabalhando em um software de edição de fotos que precisa de uma ferramenta para converter fotos coloridas para preto e branco. Implemente esta funcionalidade usando OpenCV.

### Bibliotecas/Métodos: cv2.cvtColor(), cv2.COLOR_BGR2GRAY

```python
import cv2 as cv
url = 'download.jpg'
img = cv.imread(url)
img = cv.resize(img, (0,0), fx=1, fy=1)
img = cv.cvtColor(img,cv.COLOR_BGRA2GRAY)
cv.imshow('janela',img)
cv.waitKey(0)
cv.destroyAllWindows()
```
