###  Gráfico de Barras Colorido

Situação: Uma loja deseja visualizar o desempenho de vendas mensal.
Dados: meses = ['Jan', 'Fev', 'Mar', 'Abr'], vendas = [200, 150, 300, 220].
Biblioteca: Matplotlib
Métodos: Use plt.bar(meses, vendas, color=['red', 'green', 'blue', 'yellow']) para criar barras coloridas.

```python
import matplotlib.pyplot as plt
meses = ['Jan', 'Fev', 'Mar', 'Abr']
vendas = [200, 150, 300, 220]
plt.bar(meses, vendas, color=['red', 'green', 'blue', 'yellow'])
```

### Personalizando Cores em Gráficos de Barras

Biblioteca: Matplotlib
Métodos: plt.bar(categorias, valores, color=['#aabbcc', '#ffaacc', '#ccaaff']) e plt.gcf().set_facecolor('#dddddd') para cores personalizadas.

### BeautifulSoup - Web Scraping

Situação: Extrair dados de um site de notícias para análise de tendências.
Biblioteca: BeautifulSoup
Métodos: BeautifulSoup(requests.get(url).text, 'html.parser'), .find(), .find_all()
Resolução: Use BeautifulSoup para analisar o HTML e extrair informações relevantes como títulos de notícias ou links.

```python
import requests
from bs4 import BeautifulSoup as bts

url = 'https://g1.globo.com/'

joselei = {'User - Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36' }
prr = requests.get(url, headers = joselei)


sp = bts(prr.text, 'html.parser')

sp.find_all('span')

# span class="bstn-hl-title gui-color-primary gui-color-hover gui-color-primary-bg-after"
```

### OpenPyXL - Manipulação de Excel

Situação: Ler dados de um arquivo Excel e calcular a média de vendas mensais.
Biblioteca: OpenPyXL
Métodos: load_workbook(), .active, .cell()
Resolução: Use OpenPyXL para abrir um arquivo Excel, ler valores específicos e calcular a média.

```python
from openpyxl import load_workbook

workbook = load_workbook('')
sheet = workbook.active
vendas_mensais = []
for row in sheet.iter_rows(min_row=2, min_col=2, max_col=sheet.max_column, values_only=True):
    for venda in row:
        vendas_mensais.append(venda)
media_vendas = sum(vendas_mensais) / len(vendas_mensais)
print("A média de vendas mensais é:", media_vendas)
```
